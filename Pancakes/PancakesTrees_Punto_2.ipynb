```
import pydot
import math
from collections import defaultdict 
from IPython.display import Image, display
```

```
def lstr(p):
        return ''.join([str(elem)  for elem in p])
```

```
def pope(graph,elem):
  max=0
  for i in graph[elem]:
    n=1+pope(graph,i)
    if (max<n):
      max = n
  graph.pop(elem)
  return max
```

```
def get_first_key(dictionary):
    for key in dictionary:
        return key
    raise IndexError
```

```
def ModifiedBFSOptimizedNoGraph(s): 
        # Mark the source node as visited and enqueue it 
         # Create dictionaries graph,visited and pydot graph G
         n=len(s)
         graph = defaultdict()
         visited = defaultdict()
        # Create a queue for BFS 
         queue = [] 
         queue.append(s) 
         visited[lstr(s)] = True
         while queue: 
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it
            for i in range(len(s)-1):
                  d = s[i+1::-1] + s[i+2:]
                  if (d != s) and not(lstr(d) in visited.keys()):
                    graph[lstr(s)].append(lstr(d))
                    queue.append(d) 
                    visited[lstr(d)] = True 
         file=open("output.txt","a")
         file.write("For the pancake pile of size n = "+str(len(s))+", Pn is = "+str(pope(graph,get_first_key(graph)))+" \n")
         file.close()
```

```
n = 5
while (1):
  p = []
  for i in range(n):
    p.append(i+1)
  ModifiedBFSOptimizedNoGraph(p)
  n+=1
```
